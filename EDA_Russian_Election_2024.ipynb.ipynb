```python
plt.figure(figsize=(12,6))
sns.displot(x="age",data=df,log_scale=True)
plt.xlabel("Age")
plt.ylabel("count")
plt.title("Distrubution of Age")
plt.show()
```

observation and insights
1. 35-45 has more age of peoples for election.
2. the distrubution is nearly normal distrubution.

```python
plt.figure(figsize=(12,6))
sns.histplot(x="weight1",data=df)
plt.xlabel("weight1")
plt.ylabel("Freq")
plt.title("Distrubution of weight1")
plt.show()
```

Observation and Insight

- The highest freq of weight is above 175.
- Data is right skewed data so the mean >=median>=mode

```python
plt.figure(figsize=(12,7))
ax =sns.countplot(data=df,x="federal_district",palette="magma",hue="sex")
plt.xlabel("federal district")
plt.ylabel("count")
plt.title("count of sex with respect to fedaral districts")
for i in ax.containers :
    ax.bar_label(i)

plt.show()
```

Onservation and Insight
1. There are 8 federal district for election.
2. the highest count of male and female in center federal district.

```python
## which types of city has more votes?
plt.figure(figsize=(12,7))
sns.countplot(data=df,x="type_of_city",palette="magma",hue="sex")
plt.xlabel("Type of city")
plt.ylabel("count")
plt.title("count of type of city")
plt.xticks(rotation=45)
plt.show()
```

```python
### How many peoplke has known about election date.
plt.figure(figsize=(12,9))
df.knows_election_date.value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Percentage of folk who knows election date")
plt.legend()
plt.show()
```

Observation and Insight

1. there are 4 category in data
2. As per data

- name correct date has 85.1 % folk
- not sure or no answer has 10.3% folk
- name correct year has 3.4%
- wrong answer has 1.2%
